# Explore Facebook DB

Load required packages to query SQLite database

```r
library(RSQLite) 
library(sqldf)
library(dplyr)
```

Connect to the Facebook SQLite database

```r
setwd("D:/Projects/OMTM")
db <- dbConnect(RSQLite::SQLite(), dbname="facebook.sqlite")
```

Query the 3 tables in the database

```r
posts <- dbGetQuery(db, "SELECT * FROM posts")
likes <- dbGetQuery(db, "SELECT from_id, id FROM postlikes")
comments <- dbGetQuery(db, "SELECT * FROM postcomments")
```
